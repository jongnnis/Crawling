# 과제
* 지니차트 1위 ~ 200위 까지 크롤링
* 반복문을 사용하여 크롤링

 https://www.genie.co.kr/chart/top200?ditc=D&ymd=20231122&hh=14&rtm=Y&pg=2

* 엑셀로 내보내기 기능

* 과제 제출은 github로 내보내기!


```python
import requests
from bs4 import BeautifulSoup
```

```python
title_list = []
artist_list = []

for k in range(1,5):
  url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20231122&hh=14&rtm=Y&pg={}'.format(k)
  header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
  request = requests.get(url, headers= header)
  soup = BeautifulSoup(request.text)
  for i in range(50):
    song = soup.findAll('tr',{'class':"list"})[i]
    info = song.find('td', {'class':'info'})
    title = info.findAll('a')[0].text.replace('19금\n','').strip()
    artist = info.findAll('a')[1].text.strip()
    title_list.append(title)
    artist_list.append(artist)
```

```python
for i in range(200):
  print(f'{i+1}위 {artist_list[i]} - {title_list[i]}')
```

```python
import pandas as pd
```

```python
data = []
for (i, k) in zip(artist_list, title_list):
  dic = {'artist': i, 'title': k}
  data.append(dic)

df = pd.DataFrame(data)
df
```

```python
df.to_excel('genie_music_rank.xlsx')
```
